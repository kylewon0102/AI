MSE and RMSE
-
A Loss function for regression problems that is:
- Predicting Values
- both used to calculate the difference between PREDICTED and ACTUAL


$\text{Mean Squared Error (MSE)} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
- A loss function in Regression models to optimize predictions during Training
- Measures the average of the squared errors between predicted and actual values

$\text{Root Mean Squared Error(RMSE)} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$
- Updated Version of MSE where it penalizes less heavily due to square root
- Another Common Loss Function in Regression models

```python
#Utilizing y = ax + b to perform RMSE
import numpy as np


#A basic RMSE testing code, so I gave it a and b
a = 3  #slope
b = 76 #y intercept

data = [[2, 81], [4, 93], [6, 91], [8, 97]] 
x = [i[0] for i in data] #time studied
y = [i[1] for i in data] #actual score

#predict
def predict(x):
    return a * x + b

#RMSE Calculation
def rmse(predict, actual): 
    return np.sqrt(((predict - actual) ** 2).mean())

#call RMSE
def rmse_val(predict_result, y):
    return rmse(np.array(predict_result), np.array(y))


predict_result = []
for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("Time Studied = %.f, Actual Test Score = %.f, Predicted Score = %.f" % (x[i], y[i], predict(x[i])))

print("rmse 최종값: " + str(rmse_val(predict_result, y)))
```

Gradient Descent
-
An Optimizing Strategy that:
- Utilizes derivatives (slope) to guide the direction of local minima
- Optimizer SGD and Adam utilizes gradient descent as a basis - moves based on the slope

```python
#Adding Gradient Descent into RMSE
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform

#using Function of f(x) = x^2
def f(x):
    return x**2

#d/dx(f(x)) = 2x, which is a slope
def df(x):
    return 2*x

#install x, learning rate, iterations
x = 3.0
alpha = 0.9
num_iterations = 50

x_history = [x]
f_history = [f(x)]
print(f"초기값: x = {x}, f(x) = {f(x)}") #prints 3.0 and 9

#now use for loop to adjust x value
for i in range(num_iterations):
    gradient = df(x)
    x = x - alpha * gradient
    x_history.append(x)
    f_history.append(f(x))

    print(f"Iteration {i+1}: x = {x:.4f}, gradient = {gradient:.4f}, f(x) = {f(x):.4f}")



#visualization
x_vals = np.linspace(-3, 3, 100)
y_vals = f(x_vals)
plt.figure(figsize=(7, 10))
plt.plot(x_vals, y_vals, label="f(x) = x^2")
plt.plot(x_history, f_history, '.-', label="gradient descent paths")

plt.title("Utilizing Gradient Descent to minimize x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Back to the test score predicting with RMSE and Gradient Descent
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.show()

x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0

lr = 0.01

epochs = 2001

for i in range(epochs):
    y_pred = a * x_data + b #get y pred
    #error = y_data - y_pred #get loss


    #deriv of loss func in terms of a and b. We should use the commented, but used this rather as it is binary crossentropy formula
    #it would go over if we use the commented ones
    a_diff = 2/len(x_data) * sum((a*x_data + b - y_data) * x_data)
    #a_diff = -(1/len(x_data)) * sum(x_data*(error))
    b_diff = 2/len(x_data) * sum(a*x_data + b - y_data)
    #b_diff = -(1/len(x_data)) * sum(y_data - y_pred)
    
    #gradient descent
    a = a - lr * a_diff #update a
    b = b - lr * b_diff #update b

    if i % 100 == 0: #print a and b every 100 epochs
        print(f"epoch={i}, 기울기={a}, 절편={b}")

#redrawing Graph
y_pred = a*x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)],[min(y_pred), max(y_pred)])
plt.show()
```
