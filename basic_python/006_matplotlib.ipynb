```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi*x1) *np.exp(-x1)
y2 = np.cos(2*np.pi*x1)




ax1 = plt.subplot(2, 1, 1) #2행 1열 기준 위쪽
plt.ylabel('Damped oscilation')
plt.title("2개의 서브플롯")
plt.plot(x1, y1, 'yo-')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout() #여백 감소
plt.show()
```
